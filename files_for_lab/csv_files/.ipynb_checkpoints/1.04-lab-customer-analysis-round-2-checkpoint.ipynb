```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mkt_data = pd.read_csv('marketing_customer_analysis.csv')
mkt_data
```

```python
mkt_data.shape
```

```python
print(mkt_data.columns)
```

```python
#lower cases
mkt_data.columns = mkt_data.columns.str.lower()
mkt_data
```

```python
# Standardized column names will be meaninfull strings, lower cased and with '_' instead of 'spaces'
mkt_data.columns = [mkt_data.columns[i].lower().replace(' ', '_') for i in range(len(mkt_data.columns)) ] 
mkt_data
```

```python
# Renaming columns
mkt_data = mkt_data.rename(columns={'unnamed:_0':'id'})
mkt_data
```

```python
categorical = mkt_data.select_dtypes(include='object')
categorical
```

```python
numerical = mkt_data.select_dtypes(include='number')
numerical
```

```python
#Checking NaN values
mkt_data.isnull().values.any()
```

```python
mkt_data[mkt_data['vehicle_type'].isna()==False]
```

```python
#replacing Nans in vehicle_type column with A
mkt_data['vehicle_type'] = mkt_data['vehicle_type'].fillna('A')
mkt_data['vehicle_type'].isna().sum()
```

```python
mkt_data[mkt_data['response'].isna()]
```

```python
#dropping rows that have NAs
mkt_data.dropna
```

```python
# Exporting this processed data to a csv
mkt_data.to_csv('mkt_analysis_clean_ver1.csv')
```

Datetime format
Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
mkt_analysis_clean_ver1 = pd.read_csv('mkt_analysis_clean_ver1.csv')
```

```python
mkt_analysis_clean_ver1['effective_to_date']
```

```python
mkt_analysis_clean_ver1['effective_to_date'] = pd.to_datetime(mkt_analysis_clean_ver1['effective_to_date'], errors='coerce')
mkt_analysis_clean_ver1['effective_to_date']
```

```python
mkt_analysis_clean_ver1["month"] = mkt_analysis_clean_ver1['effective_to_date'].map(lambda x: x.month)
mkt_analysis_clean_ver1.month
```
